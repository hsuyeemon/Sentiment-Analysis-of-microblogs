```python
#relate with classes from original text file
f = open('C:\\Users\\user\\Desktop\\nlp\\ark-tweet-nlp-0.3.2\\sentiment-analysis\\data\\output\\extracted.2class.txt', 'r')
lines = f.readlines()
f.close()
arr= [];
print(len(lines))
for  i  in  range ( len(lines) ): 
    arr.append(lines[i].split('\t'))
```

```python
f1 = open('C:\\Users\\user\\Desktop\\nlp\\ark-tweet-nlp-0.3.2\\sentiment-analysis\\data\\input\\topic.2class.txt', 'r')
l = f1.readlines()
f1.close()
a= [];
print(len(l))
for  i  in  range ( len(l) ): 
    a.append(l[i].split('\t'))
topics = set();
for i in range(len(arr)):
    topics.add(a[i][1]);
```

```python
file = open('C:\\Users\\user\\Desktop\\nlp\\ark-tweet-nlp-0.3.2\\sentiment-analysis\\data\\output\\mutualExclusive.2class.txt',"w") 
for  topic  in  topics: 
    #file.write(topic);
    for i in range(len(arr)):
        if a[i][1]==topic :
            file.write(arr[i][0]);            
            file.write("\t");
            file.write(arr[i][1]);
file.close();
```
