```python
#topics
f1 = open('C:\\Users\\user\\Desktop\\nlp\\ark-tweet-nlp-0.3.2\\sentiment-analysis\\data\\input\\topic.2class.txt', 'r')
l = f1.readlines()
f1.close()
a= [];
print(len(l))
for  j  in  range ( len(l) ): 
    a.append(l[j].split('\t'))

print(len(a))

cleanlist = []
for j in range (len(a)):
    if a[j][1] not in cleanlist :
        cleanlist.append(a[j][1])

print(len(cleanlist))
```
