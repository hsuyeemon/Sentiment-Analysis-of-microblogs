```python
#Read from tokenized text
f = open('E:\\JAIST-ShiraiLab\\data\\output\\tokenized.2class.txt', 'r')
#lines = f.readlines();
#.read().splitlines()
temp = f.read().splitlines()
f.close()
```

```python
#remove unnecessary ''s
temp = [x for x in temp if x != '']
temp
```

```python
#split the tokens,types,confidence and original text
tokens = []
for i in range (len(temp)):
#for i in range (3000):
  tokens.append(temp[i].split('\t'));
```

```python
tokens[0][1]
```

```python
#split the tokens and types into arrays
for i in range(len(tokens)):
    tokens[i][0]=tokens[i][0].split(" ");
    tokens[i][1]=tokens[i][1].split(" ");
```

```python
tokens[0][0]
```

```python
#remove the duplicate words in list
def unique_list(l):
    ulist = []
    [ulist.append(x) for x in l if x not in ulist]
    return ulist
```

```python
#remove the non-content words
featurePosList = {'N','O','^','S','Z','V','A','R','!','#','E'}
extractedFeatures = [];

for i in range(len(tokens)):
    #x=set();
    str="";
    for j in range(len(tokens[i][1])):
        if(tokens[i][1][j] in featurePosList):
            str+=tokens[i][0][j];
            str+=" ";
    print(str);
    str=' '.join(unique_list(str.split()))
    extractedFeatures.append(str);
    
```

```python
extractedFeatures
```

```python
#relate with classes from original text file
f = open('E:\\JAIST-ShiraiLab\\data\\input\\topic.2class.txt', 'r')
lines = f.readlines()
f.close()
arr= [];
for  i  in  range ( len(lines) ): 
    arr.append(lines[i].split('\t'))
```

```python
#write the extracted features
file = open('E:\\JAIST-ShiraiLab\\data\\output\\extracted.2class.txt',"w") 
for  i  in  range (len(temp)): 
    text = arr[i][2];
    file.write(text);
    file.write('\t');
    file.write(extractedFeatures[i]);
    file.write('\n');
    
file.close() 
```
