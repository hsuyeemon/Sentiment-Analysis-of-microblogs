```python
import numpy as np
from sklearn.model_selection import KFold
```

```python
f = open('C:\\Users\\user\\Desktop\\nlp\\ark-tweet-nlp-0.3.2\\sentiment-analysis\\data\\output\\extracted.2class.txt', 'r')
lines = f.readlines()
f.close()
arr= [];
print(len(lines))
for  i  in  range ( len(lines) ): 
    arr.append(lines[i].split('\t'))
```

```python
f1 = open('C:\\Users\\user\\Desktop\\nlp\\ark-tweet-nlp-0.3.2\\sentiment-analysis\\data\\input\\topic.2class.txt', 'r')
l = f1.readlines()
f1.close()
a= [];
#print(len(l))
for  j  in  range ( len(l) ): 
    a.append(l[j].split('\t'))
topics = set();
for k in range(len(a)):
    topics.add(a[k][1]);
#len(topics)
#print(topics)
```

```python
kf = KFold(n_splits=10)

Y=[];
import pandas as pd
df = pd.DataFrame();
for  topic  in  topics: 
    X=[];
    for i in range(len(arr)):
        if a[i][1]==topic :
            X.append(arr[i][1]);
    #Y.append(X);
    dd = pd.DataFrame(data=X);
    #print(np.transpose(dd))

    df = df.append(dd);

#df.fillna("0")
```

```python
i=0
for train, test in kf.split (df):
    # If this is the case,
    print('train index',train)
    print('test index',test)
    train_df = df.iloc [train]
    test_df = df.iloc [test]
    print ( 'train)' , train_df)
    print ( '(test)' , test_df)
    
    train_df.to_csv(r'e:\css\t'+str(i)+'.txt', header=None, index=None, sep=' ', mode='a')
    i=i+1;
```
