```python
tokens = []
#for i in range (len(lines)):
for i in range (3000):
  tokens.append(temp[i].split('\t'));
```

```python
featurePosList = {'N','O','^','S','Z','V','A','R','!','#','E'}
extractedFeatures = [];

for i in range(len(tokens)):
    #x=set();
    str="";
    for j in range(len(tokens[i][1])):
        if(tokens[i][1][j] in featurePosList):
            #x.add(tokens[i][0][j]); 
            str+=tokens[i][0][j];
            str+=" ";
            #print(x)
    #a = ' '.join(str(s) for s in x)
    #print('\n\n\n')
    #extractedFeatures.append(a);
    
    print(str);
    print('\n\n\n')
    
```
